```python
import requests
from bs4 import BeautifulSoup

page = f"https://www.unibe.edu.do/"
    
request = requests.get(page)
soup = BeautifulSoup(request.text, "html.parser")
```

```python
import re
from urllib.parse import urljoin, urlparse
from typing import List

def is_valid_url(url):
    """Check if the URL is valid"""
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)

def extract_url_webpage(soup: BeautifulSoup) -> List[str]:
    """Extract all of url from the webpage"""
    urls = []
    for tag_a in soup.find_all("a", href=True):
        url = tag_a.get("href")
        if is_valid_url(url):
            urls.append(url)
    return urls

def validate_url_webpage(list_url: List[str], base_url: str) -> List[str]:
    return list(set(list(filter(lambda url: re.search(rf"{base_url}.*", url), list_url))))
    
base_url = "https://www.unibe.edu.do"
validate_url_webpage(extract_url_webpage(soup), base_url)[:1]
```

```python
CHUNKS = 300

def extract_chunks(soup: BeautifulSoup) -> List:
    batches = []
    content = soup.text.split()
    for i in range(len(content)//CHUNKS+1):
        batches.append(" ".join(content[i*CHUNKS:(i+1)*CHUNKS]))
    return batches
```

```python
base_url = "https://www.unibe.edu.do"
urls = validate_url_webpage(extract_url_webpage(soup), base_url)[:10]
chunks = [extract_chunks(BeautifulSoup(requests.get(url).text, "html.parser")) for url in urls]

chunks
```
